```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical 


df =  pd.read_csv('AU.csv')

# split out X and y
X = df.drop(columns=['class'])
y = df[['class']].values.reshape(-1)

print("Before Encoding", len(X.columns))

X=pd.get_dummies(X)
y_one_hot=pd.get_dummies(y)
print("After Encoding", len(X.values[0]))
 

# split into testing and training
train_X,test_X,train_label,test_label = train_test_split(X, y_one_hot, test_size=0.2, random_state=13)

# scale the data by standardizing it
mean=train_X.mean(axis=0)
```

```python
mean=train_X.mean(axis=0)
train_X -=mean
std=train_X.std(axis=0)
train_X /=std
test_X -=mean
test_X /=std

# split data into training/validation
train_X,val_X,train_label,val_label = train_test_split(train_X, train_label, test_size=0.2, random_state=13)
```

```python
# B - Developing the model

import keras
from keras import layers
from keras import models

au_model= models.Sequential()
au_model.add(layers.Dense(32, activation='sigmoid',input_shape=(train_X.shape[1],)))
au_model.add(layers.Dense(64, activation='relu'))
au_model.add(layers.Dense(64, activation='relu'))
au_model.add(layers.Dense(3, activation='softmax'))

au_model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(),metrics=['accuracy'])
```

```python
# C - Reporting the training/validation

# fitting the model with training/validation data 
au_model_train = au_model.fit(train_X, train_label, batch_size=30,epochs=7,verbose=1,validation_data=(val_X, val_label))
```

```python
# C - continued...reporting the testing results and checking for overfitting

# applying the model to test data
test_eval = au_model.evaluate(test_X, test_label, verbose=0)
print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1])

# Plotting the accuracy and loss to check for overfitting
import matplotlib.pyplot as plt
%matplotlib inline

accuracy = au_model_train.history['acc']
val_accuracy = au_model_train.history['val_acc']
loss = au_model_train.history['loss']
val_loss = au_model_train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
# A - Data pre-processing - splitting, scaling, encoding

# load the data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical 


df =  pd.read_csv('AU.csv')

# split out X and y
X = df.drop(columns=['class'])
y = df[['class']].values.reshape(-1)

print("Before Encoding", len(X.columns))

 

# encode the labels and categorical variables in X
# first change to integer
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_integer_encoded = label_encoder.fit_transform(y)
X['att5'] = label_encoder.fit_transform(X['att5'])
X['att6'] = label_encoder.fit_transform(X['att6'])
X['att8'] = label_encoder.fit_transform(X['att8'])
X['att10'] = label_encoder.fit_transform(X['att10'])
print("After Encoding", len(X.values[0]))
# then change labels to one hot encoding
y_one_hot = to_categorical(y_integer_encoded)

# split into testing and training
train_X,test_X,train_label,test_label = train_test_split(X, y_one_hot, test_size=0.2, random_state=13)

# scale the data by standardizing it
mean=train_X.mean(axis=0)
```

```python
mean=train_X.mean(axis=0)
train_X -=mean
std=train_X.std(axis=0)
train_X /=std
test_X -=mean
test_X /=std

# split data into training/validation
train_X,val_X,train_label,val_label = train_test_split(train_X, train_label, test_size=0.2, random_state=13)
```

```python
# B - Developing the model

import keras
from keras import layers
from keras import models

au_model= models.Sequential()
#input layer
au_model.add(layers.Dense(32, activation='sigmoid',input_shape=(train_X.shape[1],)))
# 2 hidden layers
au_model.add(layers.Dense(64, activation='relu'))
au_model.add(layers.Dense(64, activation='relu'))
# output layer
au_model.add(layers.Dense(3, activation='softmax'))

au_model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(),metrics=['accuracy'])
```

```python
# C - Reporting the training/validation

# fitting the model with training/validation data 
au_model_train = au_model.fit(train_X, train_label, batch_size=20,epochs=7,verbose=1,validation_data=(val_X, val_label))
```

```python

# applying the model to test data
test_eval = au_model.evaluate(test_X, test_label, verbose=0)
print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1])

# Plotting the accuracy and loss to check for overfitting
import matplotlib.pyplot as plt
%matplotlib inline

accuracy = au_model_train.history['acc']
val_accuracy = au_model_train.history['val_acc']
loss = au_model_train.history['loss']
val_loss = au_model_train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```
